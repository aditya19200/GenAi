```python
from transformers import pipeline
```

```python
import torch
torch.cuda.is_available()
```

Experiment 1 — Text Generation

```python
prompt = "The future of Artificial Intelligence is"

models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}

for name, model in models.items():
    print(f"\n{name} Generation:")
    try:
        gen = pipeline("text-generation", model=model, device=0)
        print(gen(prompt, max_length=30))
    except Exception as e:
        print("Failed:", e)
```

Experiment 2 — Fill Mask (MLM)

```python
import torch
device = 0 if torch.cuda.is_available() else -1
print("Using device:", "GPU" if device == 0 else "CPU")
```

```python
mask_tokens = {
    "BERT": "[MASK]",
    "RoBERTa": "<mask>",
    "BART": "<mask>"
}

for name, model in models.items():
    print(f"\n{name} Fill-Mask:")
    try:
        text = f"The goal of Generative AI is to {mask_tokens[name]} new content."
        fill = pipeline("fill-mask", model=model, device=device)
        print(fill(text)[:3])
    except Exception as e:
        print("Failed:", e)
```

Experiment 3 — Question Answering


```python
qa_input = {
    "question": "What are the risks?",
    "context": "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
}
```

```python
for name, model in models.items():
    print(f"\n{name} QA:")
    try:
        qa = pipeline("question-answering", model=model, device=0)
        print(qa(qa_input))
    except Exception as e:
        print("Failed:", e)
```
