```python
from transformers import pipeline

# Load summarization pipeline (fast model for news TL;DR)
news_summarizer = pipeline(
    "summarization",
    model="sshleifer/distilbart-cnn-12-6"
)

# Realistic news article input (long enough for summarization)
article_text = """
The Reserve Bank of India (RBI) on Friday decided to keep the repo rate unchanged
at 6.5 percent, citing global economic uncertainty and persistent inflation
concerns. The decision was announced after the Monetary Policy Committee’s
three-day meeting and was in line with market expectations.

RBI Governor Shaktikanta Das stated that while inflation has moderated in recent
months, it remains above the central bank’s comfort level. Ongoing geopolitical
tensions, volatile crude oil prices, and slowing global growth were highlighted
as key risks to economic stability.

The central bank reiterated its commitment to maintaining a balance between
supporting economic growth and ensuring price stability. Analysts believe the
pause in rate hikes is aimed at giving previous policy measures time to transmit
through the economy while monitoring future inflation trends.
"""

# Generate TL;DR summary (3–4 sentences)
summary = news_summarizer(
    article_text,
    max_length=120,
    min_length=60,
    do_sample=False
)

print("TL;DR Summary:")
print(summary[0]['summary_text'])
```
